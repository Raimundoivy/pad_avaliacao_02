**1 – Baseados em Texto**

**CSV (Comma‑Separated Values)**

Origem/Histórico 

CSV surgiu como uma forma simples de armazenar dados tabulares em texto, muito usado em planilhas e bancos de dados desde os anos 1970–1980. Não possui um padrão único, mas a especificação mais conhecida é a RFC 4180. O CSV tem sido usado para troca e conversão de dados entre vários programas de planilha há bastante tempo. Porém, embora esse formato seja muito comum, ele nunca foi formalmente documentado.

Estrutura

Embora existam várias especificações e implementações para o formato CSV, não existe uma especificação formal, o que permite uma ampla variedade de interpretações de arquivos CSV.

- Dados organizados em linhas e colunas
- Cada registro está localizado em uma linha separada, delimitada por uma quebra de linha (CRLF)
- Campos separados por delimitador (, ou ;)

Por exemplo:

       aaa,bbb,ccc CRLF
       zzz,yyy,xxx CRLF

O último registro do arquivo pode ou não ter uma quebra de linha (CRLF) final. 

Por exemplo:
       
       aaa,bbb,ccc CRLF
       zzz,yyy,xxx

Cada campo pode ou não estar entre aspas duplas.

Por exemplo:

       "aaa","bbb","ccc" CRLF
       zzz,yyy,xxx
       
Encoding

Texto simples (UTF‑8 ou ASCII). Não há definição formal de tipos de dados: todos os valores são strings até serem convertidos pela aplicação.

Compressão

Geralmente os arquivos CSV podem ser comprimidos externamente (gzip, zip), reduzindo significativamente arquivos grandes.
Compactação binária de tabelas: formatos como Parquet ou Feather transformam CSV em binário columnar, aumentando a compactação e a velocidade de leitura.

**XML (Extensible Markup Language)**

Origem/Histórico

A Linguagem de Marcação Extensível (XML) é um formato de texto simples e muito flexível derivado do SGML (ISO 8879). XML foi criado em 1996 pelo W3C para representar dados estruturados e facilitar a troca de informações entre sistemas.

Estrutura 

Estrutura hierárquica com tags e atributos. Pense em um documento XML como a linearização de uma estrutura em árvore. Em cada nó da árvore existem várias cadeias de caracteres. A estrutura da árvore e essas cadeias de caracteres juntas formam o conteúdo de informação de um documento XML. Quase todo o restante decorre naturalmente disso. Alguns dos caracteres no documento existem apenas para dar suporte à linearização, enquanto outros fazem parte do conteúdo de informação.

Por exemplo:

<pessoa>
  <nome>Ana</nome>
  <idade>25</idade>
</pessoa>

A estrutura principal de um documento XML é semelhante a uma árvore, e grande parte da estrutura léxica é dedicada a definir essa árvore. No entanto, também existe uma forma de criar conexões entre nós arbitrários da árvore.

Por exemplo, no documento a seguir há um nó raiz com três filhos, mas um dos filhos possui um link para um dos outros filhos:
<p>
<q id="x7">The first q</q>
<q id="x8">The second q</q>
<q href="#x7">The third q</q>
</p>

Encoding 

Texto estruturado (UTF-8 ou UTF-16). Se o encoding não for informado, o padrão assumido normalmente é UTF-8.
Pode utilizar XML Schema (XSD) para validar tipos de dados.

Compressão

Minificação: Remover espaços e quebras de linha entre tags.

Compressão por algoritmo: zip, bgzip, Brotli, reduzindo arquivos XML grandes até 80–90%.

Compactação binária / especializada: EXI (Efficient XML Interchange), padrão W3C que codifica XML em formato binário, reduzindo drasticamente o tamanho e acelerando parsing.

**JSON (JavaScript Object Notation)**

Origem/Histórico

Criado por Douglas Crockford (2001) para serializar objetos JavaScript e se tornou padrão de troca de dados em APIs.
JSON (JavaScript Object Notation) é um formato leve para intercâmbio de dados. Baseia-se em um subconjunto do padrão da linguagem de programação JavaScript ECMA-262, 3ª edição, de dezembro de 1999. JSON é um formato de texto completamente independente de linguagem, mas utiliza convenções familiares a programadores da família de linguagens C, incluindo C, C++, C#, Java, JavaScript, Perl, Python e muitas outras. 

Estrutura

O JSON é construído sobre duas estruturas: uma coleção de pares nome/valor (objeto, estrutura...) e uma lista ordenada de valores. (array , vetor, lista...).
Por exemplo:

{
 "nome": "Ana",
 "idade": 25
}

Encoding

Texto estruturado (UTF-8 padrão)

Compressão

Minificação, que remove espaços, quebras de linha e indentação sem alterar os dados. >json.dumps(obj, separators=(',',':'))<

Compressão por algoritmos, como gzip ou Brotli, aplicada especialmente em APIs e transferência de dados na web.

Redução de nomes de chaves ou uso de formatos binários como MessagePack e CBOR, que representam os mesmos dados de forma mais compacta.
